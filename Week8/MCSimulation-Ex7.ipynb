(a)

$f(x) = 60x^3(1-x)^2, 0 \le x\le1$

```python
x = arange(0.0, 1.0+0.001,0.001)
f = lambda x: 60*(x**3)*((1-x)**2)
xmin = 0
xmax = 1
subx = x[logical_and((x>xmin),(x<xmax))]#computes truth of xmin and xmax
ymax= max(f(subx))*1.05
ymin=0

N=10000
samples_x = xmin + (xmax-xmin)*random_sample(N)
samples_y = ymin + (ymax-ymin)*random_sample(N)

plot(x,f(x),lw=3)
plot(samples_x[1:200], samples_y[1:200], '.')
plot([xmin, xmin], [ymin, ymax], color='k', linestyle='-')
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='-')
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='-')
show()
```

```python
newmask = (samples_y < f(samples_x))
plot(x,f(x),lw=3)
plot(samples_x[1:200], samples_y[1:200], '.')
plot([xmin, xmin], [ymin, ymax], color='k', linestyle='-')
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='-')
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='-')
plot(samples_x[newmask[0:200]], samples_y[newmask[0:200]], 'r.')
```

(b)

$F(x)=15x^4-24x^5+10x^6$

```python
F = lambda x: (15*(x**4))-(24*(x**5))+(10*(x**6))
distribution = samples_x[newmask]
plot(x,f(x),'r-', lw=3, label='Density distribution')
plot(x,F(x),'g--', lw=3, label='CDF')
hist(distribution,50,normed=True, label='Sampled distribution')
legend(loc=0)
show()
```
