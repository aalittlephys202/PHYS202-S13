create a very simple plot

```python
t = arange(0.0, 1.0+0.01, 0.01)
s =cos(2*2*pi*t)
c =sin(2*2*pi*t)
```

```python
fig = plot(t, s, color='blue',linewidth=2.5, linestyle= ":",label="$cos(2*2*pi*t)$")
plot(t, c, color='red',linewidth=2.5, label="$sin(2*2*pi*t)$")
xlabel("time(s)", fontsize=20)
ylabel('volts(mV)', fontsize=20)
title('A simple plot',fontsize=20)
grid(True)
savefig('simple plot.pdf')
legend(loc="lower left")
```
