==========

```python
k = 0.0**2
k1 = 0.5**2
k2 = 1.0**2
k3 = 2.0**2


t = arange(-pi, pi +0.01,0.01)
U = -(1/cosh(t))**2
a = (k + (tanh(t))**2)/(1+k)
b = (k1 + (tanh(t))**2)/(1+k1)
c = (k2 + (tanh(t))**2)/(1+k2)
d = (k3 + (tanh(t))**2)/(1+k3)
xlabel("x(rad)")
ylabel("U(x),psi^2(x)")
grid(True)
plot(t,U, label="U(x)")
plot(t,a, label="k=0.0", color="red")
plot(t,b, label="k=0.5", linestyle=":", color="red")
plot(t,c, label="k=1.0", linestyle=":", color="red")
plot(t,d, label="k=2.0", linestyle=":", color="red")
legend(loc="lower left")
xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
```

```python
data = loadtxt('/Users/Alex/Desktop/sunspots.txt',float)
x = data[:,0]
y = data[:,1]
a = np.zeros(1000)
b = np.zeros(1000)
avg = np.zeros(1000)
tot = 0
for i in range(1000):
    a[i] = x[i]
    b[i] = y[i]
    tot += y[i]
    avg[i] = tot / i

    


plot(a, b, lw = 1, label = "Sunspots")
plot(a, avg, lw = 5, label = "Average")
xlabel("Months")
ylabel("# of Sunspots")
title("First 1000 Sunspots since 1749")
legend()
```
