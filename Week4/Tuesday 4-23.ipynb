Electric Dipole Potential
=========================

\begin{equation}
V(x,y)=\frac{kq}{\sqrt{(x^2)+(y-d)^2}}-\frac{kq}{\sqrt{(x^2)+(y+d)^2}}
\end{equation}

```python
k=1 ; q=1 ;d=1
```

```python
V = (k*q/(x**2 + (y - d)**2)**(1/2.) - (k*q/(x**2 + (y+d)**2)**(1/2.)))
G = exp((-(x**2)-(y**2)))
#put a period after 1/2  because it was a float(something about float)
print V
```
