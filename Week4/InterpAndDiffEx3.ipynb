```python
def finiteDifference(x,y):
    """returns derivative"""
    dydx= zeros(y.shape, float)
    dydx[1:-1] = (y[2:]-y[:-2])/(x[2:]-x[:-2])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

(b)
===

```python
x = arange(0.,10.,0.1)
y = x**5
z= finiteDifference(x,y)
print z
```

```python
x = arange(0,10.,0.1)
print z-(5*(x**4))
```

```python
def fourPtFiniteDiff(x,y):
    """this function will return the derivative of y with respect to x with x and y as arrays using the four-point center differencing method"""
    dydx = zeros(y.shape, float)
    dydx[2:-2] = (y[0:-4]-(8*y[1:-3])+(8*y[3:-1])-y[4:])/(12*(x[3]-x[2]))
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[1] = (y[2]-y[1])/(x[2]-x[1])
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
x = arange(0.,5, 0.1)
y = x**5
print fourPtFiniteDiff(x,y)
```

```python
Realdydx = 5*x**4
plot(x, Realdydx, linestyle=':', linewidth=7, color='blue', label='True Derivative')
plot(x, fourPtFiniteDiff(x,y), linewidth=3, color='red', label='part (a) difference')
legend(loc='upper left')
```

```python
fourPtDiff = fourPtFiniteDiff(x,y)
percentDiff= 100*abs(fourPtDiff[1:]-Realdydx[1:])/Realdydx[1:]
semilogy(x[1:], percentDiff, label='Percent Error')
legend(loc='upper right')
show()
```
