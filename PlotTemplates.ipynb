Publication ready plot template
===============================

Simple graphing template for matplotlib functions and data plots

Line Plots

```python
y = arange(-5.0, 6.0)
E = cosh(y)
p = sinh(y)
print E
print p
```

```python
# I defined my y to be between values -5.0 and 6.0. Then set E to its equation and p to its equation
```

(d)

```python
plot(y,E)
plot(y,p)
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['top'].set_position(('data', 0))
ax.spines['right'].set_position(('data', -1))
```

Subplots

```python
#Plotting sinh and cosh
#subplot(m,n,P) creates an MxN matrix, and places the following plot in the P position 
subplot(1,2,1); plot(y,E)
title('Cosh')
subplot(1,2,2); plot(y,p)
title('Sinh')
 
#ax = gca needed to modify axes
ax = gca()
#deleting boundaries
ax.spines['right'].set_color('none')

ax.spines['top'].set_color('none')

 

 

ax.spines['left'].set_position(('data', 0))

ax.spines['bottom'].set_position(('data', 0))

ax.spines['top'].set_position(('data', 0))

ax.spines['right'].set_position(('data', -1))
```
